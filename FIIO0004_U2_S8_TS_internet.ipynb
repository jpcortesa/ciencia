## 1. Carga de datos desde internet
Se carga directamente el dataset desde una URL pública, cumpliendo con el requerimiento del taller.

```
import pandas as pd

# URL directa del dataset (sin descarga local)
url = "https://raw.githubusercontent.com/yousefhossam2222/customer-churn-dataset-csv/main/Customer%20Churn%20Dataset.csv"

df = pd.read_csv(url)

# Verificar carga correcta
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])

df.head()
```

### Inspección inicial

```
df.info()
```

```
df.isnull().sum()
```

```
df.describe()
```

## 2. Limpieza y transformación de datos

```
# Eliminar duplicados y nulos
df = df.drop_duplicates()
df = df.dropna()
```

### Discretización

```
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[0, 25, 40, 60, 100],
                         labels=['Joven', 'Adulto', 'Maduro', 'Senior'])

df['Tenure_Group'] = pd.cut(df['Tenure'],
                            bins=[0, 12, 36, 60, 100],
                            labels=['Nuevo', 'Medio', 'Estable', 'Leal'])

df['Spend_Group'] = pd.cut(df['Total Spend'],
                           bins=3,
                           labels=['Bajo', 'Medio', 'Alto'])
```

### One-Hot Encoding

```
df_encoded = pd.get_dummies(df[['Gender',
                                'Usage Frequency',
                                'Age_Group',
                                'Tenure_Group',
                                'Spend_Group',
                                'Churn']])
```

## 3. Aplicación del algoritmo Apriori

```
from mlxtend.frequent_patterns import apriori, association_rules

frequent_items = apriori(df_encoded, min_support=0.05, use_colnames=True)

rules = association_rules(frequent_items, metric="confidence", min_threshold=0.6)

rules = rules.sort_values(by='lift', ascending=False)

rules.head()
```

## 4. Visualización

```
import matplotlib.pyplot as plt

top_rules = rules.head(10)

plt.figure()
plt.barh(range(len(top_rules)), top_rules['lift'])
plt.yticks(range(len(top_rules)), top_rules.index)
plt.xlabel("Lift")
plt.title("Top 10 Reglas por Lift")
plt.show()
```

## 5. Interpretación y conclusiones

- Clientes nuevos y de bajo uso presentan mayor churn.
- Clientes leales y con alto gasto muestran mayor fidelidad.
